## Binary **Conversion**

```python
varlist =  ['Gender']
def binary_map(x):
    return x.map({'Male': 1, "Female": 0})

onlineads[varlist] = onlineads[varlist].apply(binary_map)

# Check the housing dataframe now
onlineads
```

## **Categorical Conversion**

```python
# Dropping the first column from status dataset
status = pd.get_dummies(onlineads['Social Media'], drop_first = True)

# Adding the status to the original housing dataframe
onlineads = pd.concat([onlineads, status], axis = 1)

# Dropping 'furnishingstatus' as we have created the dummies for it
onlineads.drop(['Social Media'], axis = 1, inplace = True)

onlineads
```

# **outlair**

```python
def detect_outliers_zscore(data):
    outliers = []
    thres = 1
    mean = np.mean(data)
    std = np.std(data)
  
    for i in data:
    
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers
```

```python
sample_outliers_Age = detect_outliers_zscore(onlineads['Age']) 
print("Outliers in Age column from Z-scores method: ", sample_outliers_Age)
```

```python
from scipy.stats import zscore
z1=np.abs(zscore(onlineads['Age']))
onlineads['Zscore_Age']=z1
outliers_Age=len(onlineads[onlineads['Zscore_Age']>3])
data_z = onlineads[onlineads['Zscore_Age']<3]
onlineads=data_z.drop(['Zscore_Age'], axis=1)
onlineads
```
